```python
from nltk.corpus import brown
print(brown.categories())
news_text=brown.words(categories='news')#获取词
# print(news_text)
# print(brown.sents(categories='news'))#获取句子

import nltk
fdist=nltk.FreqDist(w.lower() for w in news_text)#频率分布
print(fdist['people'])
searchWords=["people","life","love"]
#用循环输出这三个词在新闻语料中词频
for m in searchWords:
    print(m+":",fdist[m])
```

```python
#条件频率分布
import nltk
from nltk.corpus import brown
observeWords=["can","could","may","might","must","will"]
genre_word=[(genre,word) for genre in ["news","romance"]
            for word in brown.words(categories=genre)]
print(genre_word[:4])
print(genre_word[-4:])
cfd=nltk.ConditionalFreqDist(genre_word)
cfd.tabulate(samples=observeWords)
%matplotlib inline
cfd.plot(samples=observeWords)
```

```python
import nltk
from nltk.corpus import brown
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
genre_word=[(genre,word) for genre in ["news","romance"]
            for word in brown.words(categories=genre)]
print(genre_word[:4])
print(genre_word[-4:])
cfd=nltk.ConditionalFreqDist(genre_word)
cfd.tabulate(samples=days)
%matplotlib inline
cfd.plot(samples=observeWords)
```

```python
#计算某个字符串，例如"tomorrow son"，包含的字节数
from nltk.corpus import cmudict
prodict=cmudict.dict()#字典 词：发音
sum=0
for word in ["tomorrow","son"]:
    print(prodict[word])
    sum=sum+len(prodict[word][0])
print(sum)
```

```python
#WordNet-查找某个词的左右语义，例如motorcar
from nltk.corpus import wordnet as wn
print(wn.synsets('motorcar'))#义源。词性。编号
print(wn.synset('car.n.01').lemmas())#相关的语义
print(wn.synsets('motorcar')[0].lemmas())#相关的语义
print(wn.synsets('motorcar')[0].lemma_names())
#获取car的相关lemma
for synsets in wn.synsets('car'):
    print(synsets.lemma_names())
```

```python
#反义关系
print(wn.lemma('supply.n.02.supply').antonyms())#词义的反义
print(wn.lemma('rush.v.01.rush').antonyms())

#计算语义相似度，right_whale与minke_whale
right_whale=wn.synsets('right_whale')[0]
minke_whale=wn.synsets('minke_whale')[0]
right_whale.path_similarity(minke_whale)

#计算上位词下位词
dog=wn.synset('dog.n.01')
print(dog.hypernyms())#上位词
print(dog.hyponyms())#下位词
print(dog.member_holonyms())#整体
print(wn.synset('tree.n.01').member_holonyms())#整体
print(wn.synset('dog.n.01').lowest_common_hypernyms(wn.synset('cat.n.01')))#共同上位词
```
